```python
#Here in the steps below we are importing pandas library , train_test_split to split our data set for training and for 
#testing , confusion matrix to check if the the predicted output is aligning with the original value
#DecisionTree classifier for using decision tree.
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.preprocessing import LabelEncoder


df = pd.read_csv("PersonData.csv")

#
gender_encoder = LabelEncoder()

#since we have gender values as male and female we have to encode it.
df['Gender'] = gender_encoder.fit_transform(df['Gender'])


#Hieght weight and footsize will be input and gender will be output.
X = df[['Height', 'Weight', 'Foot_Size']]
y = df['Gender']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


classifier = DecisionTreeClassifier()


classifier.fit(X_train, y_train)


height = float(input("Enter height: "))
weight = int(input("Enter weight: "))
foot_size = int(input("Enter foot size: "))
new_entry = [[height, weight, foot_size]]


prediction = classifier.predict(new_entry)

#Now decoding numerical out of 0 and 1 to male or female.
predicted_output = gender_encoder.inverse_transform(prediction)

# Display the prediction for the new entry
print(f"Prediction for new entry with Height={height}, Weight={weight}, Foot_Size={foot_size}: {predicted_output[0]}")

# Add the new entry to the dataset as a new row
df = pd.concat([df, pd.DataFrame({'Gender': prediction, 'Height': [height], 'Weight': [weight], 'Foot_Size': [foot_size]})], ignore_index=True)

# Re-train the model on the updated dataset
X = df[['Height', 'Weight', 'Foot_Size']]
y = df['Gender']
classifier.fit(X, y)

# Predict on the test set
y_pred = classifier.predict(X_test)

# Calculate confusion matrix and accuracy
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")

# Decode Gender column back to original labels and save the updated dataset to the same CSV file
df['Gender'] = gender_encoder.inverse_transform(df['Gender'])
df.to_csv("PersonData.csv", index=False)
```
